```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

team_data = pd.read_csv('../data/team_data.csv')
game_data = pd.read_csv('../data/game_data.csv')
```

```python
if 'TEAM_NAME' in team_data.columns and 'W' in team_data.columns and 'L' in team_data.columns:
    team_data['Win_Percentage'] = team_data['W'] / (team_data['W'] + team_data['L'])
    win_percentage = team_data.groupby('TEAM_NAME')['Win_Percentage'].mean().sort_values()
    win_percentage.plot(kind='barh', figsize=(10, 8), color='gold')
    plt.title("Win Percentage by Team")
    plt.xlabel("Win Percentage")
    plt.ylabel("Team Name")
    plt.show()
```

```python
if 'TEAM_NAME' in team_data.columns and 'FG_PCT' in team_data.columns:
    team_avg_fg_pct = team_data.groupby('TEAM_NAME')['FG_PCT'].mean().sort_values()
    team_avg_fg_pct.plot(kind='barh', figsize=(10, 8), color='teal')
    plt.title("Average Field Goal Percentage by Team")
    plt.xlabel("Field Goal Percentage")
    plt.ylabel("Team Name")
    plt.show()
```

```python
if 'AST' in team_data.columns and 'TOV' in team_data.columns:
    team_data['AST_TOV_Ratio'] = team_data['AST'] / team_data['TOV']
    ast_tov_ratio = team_data.groupby('TEAM_NAME')['AST_TOV_Ratio'].mean().sort_values()
    ast_tov_ratio.plot(kind='barh', figsize=(10, 8), color='brown')
    plt.title("Assist-to-Turnover Ratio by Team")
    plt.xlabel("Assist-to-Turnover Ratio")
    plt.ylabel("Team Name")
    plt.show()
```
